## Gaussian Algebra
Generate 100000 samples each from distributions $X=\mathcal{N}(0,1)$ and $Y=\mathcal{N}(0, 2^2)$.

The `randn` function allows us to sample numbers from the normal distribution. To vary the mean $mu =$`mu` and variance $\sigma^2 =$`var` in order to generate the distribution $\mathcal{N}(\mu, \sigma^2)$ run

```
sqrt(var) * randn() + mu
```

````octave
n = 100000;
x = randn(n, 1);
y = 2 * randn(n, 1);
````

We will examine the results of addition, multiplication and division of Gaussian distributions by defining
$$\begin{align*}
z_1 &= x + y\\
z_2 &= x * y\\
z_3 &= x / y
\end{align*}
$$
### Gaussian Addition
The result of the addition of two Gaussian samples is still Gaussian. In fact, linear combinations of them are Gaussian!

````octave
z1 = x + y;
histfit(z1);
xlim([-10 10])
````

### Gaussian Multiplication
The result of Gaussian multiplication is not Gaussian. Since our normal distributions were centred around 0, multiplication by small numbers leads to even smaller numbers leading to an abundant number of products near 0.

````octave
z2 = x .* y;
histfit(z2);
xlim([-10 10])
````

### Gaussian Divison
The result of Gaussian division is not Gaussian. 

````octave
z3 = x ./ y;
histfit(z3);
xlim([-200 200])
````

## Gaussian Linear Combination
As mentioned above, a linear combination of Gaussian distributions results in a Gaussian. We can see this by plotting a histogram of the linear combination $$\begin{bmatrix}0.2 & 0.3\end{bmatrix} \begin{bmatrix}x\\ y\end{bmatrix} = 0.2 x + 0.3y$$

````octave
histfit(0.2 * x + 0.3 * y)
````
