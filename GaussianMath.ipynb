# Gaussian Math
Generate 100000 samples each from distributions $X=\mathcal{N}(0,1)$ and $Y=\mathcal{N}(0, 1)$.

The `randn` function allows us to sample numbers from the normal distribution.

```octave
n = 100000;
x = randn(n, 1);
y = 2 * randn(n, 1);
```

We will examine the results of addition, multiplication and division of Gaussian distributions by defining
$$z_1 = x + y$$
$$z_2 = x * y$$
$$z_3 = x / y$$
##Gaussian Addition
The result of the addition of two Gaussian samples is still Gaussian. In fact, linear combinations of them are Gaussian!

```octave
z1 = x + y;
figure(1);clf;hold on;
histfit(z1);
xlim([-10 10])
```

##Gaussian Multiplication
The result of Gaussian multiplication is not Gaussian. Since our normal distributions were centred around 0, multiplication by small numbers leads to even smaller numbers leading to an abundant number of products near 0.

```octave
z2 = x .* y;
histfit(z2);
xlim([-10 10])
```

## Gaussian Divison
The result of Gaussian division is not Gaussian. 

```octave
z3 = x ./ y;
histfit(z3);
xlim([-200 200])
```
